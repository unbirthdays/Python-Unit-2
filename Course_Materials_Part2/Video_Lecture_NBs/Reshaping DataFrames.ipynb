## Limits of pivot()

```python
table2 = pd.read_csv("table2.csv")
```

```python
table2.head(20)
```

```python
table2.info()
```

```python
table2.shape
```

```python
table2.pivot(index = "Country", columns = "Medal", values = "Count")
```

this is because of the year column having different values but not being taken into account by the pivot. pivot alone does not do aggregations.

cannot make a multi-index with pivot either

```python
table2.groupby(["Country", "Year", "Medal"]).Count.sum()
```

```python
table2_piv = table2.groupby(["Country", "Year", "Medal"]).Count.sum().unstack().fillna(0)
```

```python
table2_piv
```

```python
table2_piv.info()
```

```python
table2_piv.shape
```

## pivot_table()

```python
table2.pivot_table(index = "Country", columns = "Medal", values = "Count", aggfunc = "sum", fill_value = 0) # adding margins = True as an argument will add an All column to show the sums of rows and columns
```

pivot_table is used for aggregations! aggfunc will fill the data with what you want that makes sense.

```python
table2.groupby(["Country", "Year", "Medal"]).Count.sum().unstack(fill_value = 0) 
```

the combination of the groupby method and unstack method, it's the same as using the pivot_table method!

```python
agg = table2.groupby(["Country", "Year", "Medal"]).Count.sum().unstack(fill_value = 0)
```

```python
agg.head()
```

```python
agg["All"] = agg.sum(axis = 1)
```

```python
agg
```

```python
agg.loc["All"] = agg.sum(axis = 0)
```

```python
agg
```

technically we can manually add these All columns anyway

## pd.crosstab()

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
pd.crosstab(titanic.sex, titanic.pclass)
```

```python
pd.crosstab(titanic.sex, titanic.pclass, margins = True)
```

```python
pd.crosstab(titanic.sex, titanic.pclass, margins = True, normalize = True)
```

getting relative quantities

```python
titanic.groupby(["sex", "pclass"]).pclass.count().unstack()
```

as factors, you can pass two columns of two different dataframes vs the groupby method only using one dataframe.cross tabulations with frequencies should use the crosstab function.

```python
pd.crosstab(index = titanic.sex, columns = titanic.pclass, values = titanic.age, aggfunc = "mean") 
```

```python
titanic.pivot_table(index = "sex", columns = "pclass", values = "age", aggfunc = "mean")
```

crosstab has the same functionalities as pivot_table but can use multiple dataframes

```python
titanic.groupby(["sex", "pclass"]).age.mean().unstack()
```

all accomplishes the same thing

```python
pd.crosstab(index = titanic.sex, columns = titanic.pclass, values = titanic.fare, aggfunc = "sum", margins = True)
```

```python
pd.crosstab(index = titanic.sex, columns = titanic.pclass, values = titanic.fare, aggfunc = "sum", margins = True, normalize = True)
```

another real quantities vs relative quantities change using the normalize argument

```python
summer = pd.read_csv("summer.csv")
```

```python
summer
```

```python
summer.info()
```

```python
pd.crosstab(index = [summer.Year, summer.Country], columns = summer.Medal, aggfunc = "count", values = summer.Athlete, margins = True).fillna(0)
```

```python
summer.pivot_table(index = ["Year", "Country"], columns = "Medal", aggfunc = "count", values = "Athlete", fill_value = 0, margins = True)
```

```python
summer.groupby(["Year", "Country", "Medal"]).Medal.count().unstack(fill_value = 0)
```

these are all advanced aggregations!

## Melting DataFrames with melt()

```python
table_2012 = pd.read_csv("table_2012.csv")
```

```python
table_2012
```

```python
table_2012.shape
```

```python
table_2012.melt(id_vars = "Country", value_vars = ["Gold", "Silver", "Bronze"])
```

melt will unpivot a dataframe to long format

```python
melt_2012 = table_2012.melt(id_vars = "Country", value_vars = ["Gold", "Silver", "Bronze"], var_name = "Medal", value_name = "Count")
```

```python
melt_2012
```

```python
melt_2012.pivot(index = "Country", columns = "Medal", values = "Count")#.sort_values("Gold", ascending = False)
```

```python
melt_2012.pivot(index = "Country", columns = "Medal", values = "Count").sort_values("Gold", ascending = False)
```

shows who has the most gold medals first

melt and pivot are the opposites of each other
