### Step 2: Choose Method to Re-Express Variables

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
house_data.view.unique()
```

```python
house_data.house_condition.unique()
```

We want to use numbers for machine learning algorithms to be able to process the data.
For the waterfront column, there are only "Yes" and "No" values that can easily be booleans of Yes = 1, No = 0.
For the view column, we have 5 values:
- Very Poor = 1
- Poor = 2
- Neutral = 3
- Good = 4
- Great = 5 
For the house_condition column, we have 5 values:
- Very Bad Condition = 1
- Bad Condition = 2
- OK Condition = 3
- Good Condition = 4
- GreatCondition = 5 

### Step 3: Perform Transformation 

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
house_data['waterfront'] = house_data['waterfront'].replace({'Yes' : 1, "No" : 0})
```

```python
house_data['view'] = house_data['view'].replace({'Very Poor' : 1, 'Poor' : 2, 'Neutral' : 3, 'Good' : 4, 'Great' : 5})
```

```python
house_data['house_condition'] = house_data['house_condition'].replace({'Very Bad Condition' : 1, 'Bad Condition' : 2, 'OK Condition' : 3, 'Good Condition' : 4, 'Great Condition' : 5})
```
