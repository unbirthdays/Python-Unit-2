# Filtering DataFrames by One Condition

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head(10)
```

```python
titanic.sex.head(10)
```

```python
titanic.sex == "male"
```

```python
titanic[titanic.sex == "male"]
```

```python
titanic.loc[titanic.sex == "male"]
```

use loc notation as best practice

```python
titanic.loc[titanic.sex == "male", "fare"]
```

```python
titanic.loc[titanic.sex == "male", "fare"].describe()
```

```python
mask1 = titanic.sex == "male"
mask1
```

```python
titanic_male = titanic.loc[mask1]
```

```python
titanic_male.head()
```

```python
titanic.dtypes #== object
```

```python
mask2 = titanic.dtypes == object
mask2
```

```python
titanic.loc[:, ~mask2]
```

~mask2 means opposite of mask2, aka all columns with non string objects (numeric)

```python
titanic.loc[mask1, ~mask2]
```

if you filter dataframes based on a condition in the columns, use loc notation

# Filtering DataFrames with Many Conditions (using OR)

```python
titanic.head()
```

```python
mask1 = titanic.sex == "female"
mask1.head(20)
```

```python
mask2 = titanic.age < 14
mask2.head(20)
```

```python
(mask1 | mask2).head(11)
```

2 false conditions will return a false, any true condition will return a true. counts both females or anyone under 14

```python
(mask1 | mask2).head(20)
```

```python
titanic.loc[mask1 | mask2]
```

```python
wom_or_chi = titanic.loc[mask1 | mask2, ["survived", "pclass", "sex", "age"]]
```

```python
wom_or_chi.head()
```

```python
wom_or_chi.info()
```

```python
wom_or_chi.describe()
```

women or children survived more than men

```python
og_1988.head()
```

```python
summer.Year.between(1960, 1969).head()
```

```python
og_60s = summer.loc[summer.Year.between(1960, 1969, inclusive = "both")]
og_60s
```

```python
og_60s.head()
```

```python
og_60s.tail()
```

```python
my_favorite_games = [1972, 1996]
```

```python
summer.Year.isin(my_favorite_games).head()
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_favorite_games)]
```

```python
og_72_96.head()
```

```python
og_72_96.tail()
```

```python
og_not_72_96 = summer.loc[~summer.Year.isin(my_favorite_games)]
```

```python
og_not_72_96
```

```python
og_72_96.head()
```

```python
og_not_72_96.Year.unique()
```
