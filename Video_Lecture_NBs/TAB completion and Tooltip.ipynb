# Make it easy: TAB Completion and Tooltip

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic
```

```python
titanic.sort_values("age")
```

SHIFT + TAB will show all values to sort by

SHIRT + TAB + TAB will extend down for scrolling

```python
titanic.sort_values(by = "age", ascending= True)
```

```python
titanic.sort_values(by = "age", ascending= False)
```
